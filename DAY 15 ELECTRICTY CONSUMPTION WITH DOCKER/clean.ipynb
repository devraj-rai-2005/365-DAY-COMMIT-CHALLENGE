```python
import pandas as pd
```

```python
df = pd.read_csv('household_power_consumption.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
df.drop('index' , axis =1 , inplace = True )
```

```python
df.head()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['second'] = df['Time'].dt.second
df['minute'] = df['Time'].dt.minute
df['hour'] = df['Time'].dt.hour
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['day'] = df['Date'].dt.day
df['month'] = df['Date'].dt.month
```

```python
df.drop(['Date' , 'Time'] , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df['month'].value_counts().sort_index()
```

```python
df.columns = ['Global_active_power', 'Global_reactive_power', 'Voltage',
       'Global_intensity', 'kitchen', 'laundry_room',
       'HVAC', 'second', 'minute', 'hour', 'day', 'month']
```

```python
df.tail()
```

```python
df['Global_active_power']  = df['Global_active_power'].astype(float)
df['Global_reactive_power']  = df['Global_reactive_power'].astype(float)
df['Voltage']  = df['Voltage'].astype(float)
df['Global_intensity']  = df['Global_intensity'].astype(float)
df['kitchen']  = df['kitchen'].astype(float)
df['laundry_room']  = df['laundry_room'].astype(float)
```

```python
df.info()
```

```python
new_df = df.copy()
```

```python
new_df.info()
```

```python
new_df.dropna(inplace = True)
```

```python
new_df.shape
```

```python
new_df.to_csv('clean_energy_consumption.csv' , index = False)
```
