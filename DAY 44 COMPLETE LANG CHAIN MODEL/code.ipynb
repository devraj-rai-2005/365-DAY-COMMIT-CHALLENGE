```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
from pydantic import BaseModel , Field

class movie(BaseModel):
    song : str = Field(description= 'Write Song Name you Remind or a lyrics')
```

```python
from langchain_core.prompts import PromptTemplate 
from langchain_core.output_parsers import PydanticOutputParser

parser = PydanticOutputParser(pydantic_object=movie)


template = PromptTemplate(
    template='Give me a movie name from the given lyrics{song_name} \n {format_instruction}' , input_variables=['song_name'] , 
    partial_variables= {'format_instruction' : parser.get_format_instructions()})
```

```python
chain = template | model | parser 

result = chain.invoke({'song_name' : 'aye ho meri jindagi me tum bahar ban ke'})
```
