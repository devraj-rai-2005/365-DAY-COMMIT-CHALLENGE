```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel , Field 
from typing import Literal , Optional
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
load_dotenv()

model = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
class person(BaseModel):

    Name : Optional[str] = Field(description='Extract the name of the peroson in review')

    Age : Optional[int] = Field(description='Extract the Age from the given review ')

    City : Optional[str] = Field(description='Extract the city name from the given review ')
    
    sentiment : Literal['Positive' , 'Negative' , 'Neutral'] = Field(description='Check the Sentiment of the given review')

    key_points : list[str] = Field(description= 'Give me 3 best key points from the given review') 

    summary : str = Field(description='Generate the short summary of the given review')
```

```python
parser = PydanticOutputParser(pydantic_object=person)
```

```python
prompt = PromptTemplate(
    template='Extract Name , Age  , City , Sentiment , Key points and a Summary from the given following text \n {text} \n {format_instruction}' , 
    input_variables=['text'] , 
    partial_variables={'format_instruction' : parser.get_format_instructions()}
)
```

```python
text = '''This SmartPhone is very bad , i don't like it , it is a worst product from the samsung as a s25 Ultra and i don't suggest people to buy this product '''
```

```python
chain = prompt | model | parser 

result = chain.invoke({'text' : text})
```

```python
print(result)
print(result.Name)
print(result.Age)
print(result.City)
print(result.sentiment)
print(result.key_points)
print(result.summary)
```
