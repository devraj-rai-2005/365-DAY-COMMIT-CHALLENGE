```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableParallel
```

```python
load_dotenv()

model = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
parser = StrOutputParser()
```

```python
prompt_1 = PromptTemplate(
    template='Extract the name from following given review, and in case if not given take any Russian Girl Name  \n {text}' , 
    input_variables=['text']
)



prompt_2 = PromptTemplate(
    template='Extract the Age from following given review, if not given take any age under 70 \n {text}' , 
    input_variables=['text']
)



prompt_3 = PromptTemplate(
    template='Extract the city from following given review \n {text}' , 
    input_variables=['text']
)



prompt_4 = PromptTemplate(
    template='Extract the sentiment from following given review \n {text}' , 
    input_variables=['text']
)

prompt_5 = PromptTemplate(
    template='Extract the 3 main key point from following given review \n {text}' , 
    input_variables=['text']
)



prompt_6 = PromptTemplate(
    template='Generate the summary from following given review \n {text}' , 
    input_variables=['text']
)
```

```python
parallel_chain = RunnableParallel({
    'name' : prompt_1 | model | parser , 
    'age' : prompt_2 | model | parser , 
    'city' : prompt_3 | model | parser , 
    'sentiment' : prompt_4 | model | parser , 
    'key_point' : prompt_5 | model | parser , 
    'summary' : prompt_6 | model | parser , 
})
```

```python
prompt_7 = PromptTemplate(
    template='Merge all the document in the single document for taking the documents name ->{name} , age- > {age} , city -> {city} , sentiment -> {sentiment} ,  key_point -> {key_point} ,  summary -> {summary} ' , 
    input_variables=['name' , 'age' , 'city', 'sentiment', 'key_points',  'summary']
)
```

```python
merger_chain =  prompt_7 | model | parser
```

```python
review_text = '''The Face wash is very great i a dam happy to have this product in my shell and the morning of Calfornia is great and it makes my pimple and a dark spot disappear from my face.'''
```

```python
chain = parallel_chain  | merger_chain

result = chain.invoke({'text' : review_text})
```
