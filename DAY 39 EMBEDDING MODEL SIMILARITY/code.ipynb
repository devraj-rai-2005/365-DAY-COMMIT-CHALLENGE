```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
load_dotenv() 

embedding_model = GoogleGenerativeAIEmbeddings(model='models/gemini-embedding-001',output_dimensionality=300)
```

```python
document = [
    "Machine Learning (ML)is A critical subset that allows systems to learn from data, identify patterns, and improve performance without explicit programming. It is subdivided into supervised learning, unsupervised learning, and reinforcement learning." , 

    "Deep Learning (DL)is A specialized subset of machine learning that utilizes multi-layered artificial neural networks to mimic the human brain's decision-making process. It is the primary technology behind complex tasks like image and speech recognition." , 

    "Natural Language Processing (NLP) is Enables computers to understand, interpret, and generate human language, both in written and spoken form. Applications include chatbots, virtual assistants (Siri, Alexa), and translation services." , 

    "Computer Vision is Enables machines to analyze and interpret visual information from the world, such as images and videos. It is crucial for applications like facial recognition and object detection in self-driving cars." , 

    "Robotics is A branch of AI and engineering focused on designing, creating, and operating robots that can perform tasks automatically or semi-automatically." , 

    "Generative AI is A subset of deep learning focused on creating new, unique content (text, images, music) based on patterns learned from existing data"
]

doc_embedding = embedding_model.embed_documents(document)

print(len(doc_embedding[0]))
```

```python
query = "tell me about Machine Learning"

query_embedding = embedding_model.embed_query(query)

print(len(query_embedding))
```

```python
score_list = cosine_similarity([query_embedding] , doc_embedding).flatten()
```

```python
index = np.argmax(score_list)
```

```python
score  = np.round(score_list[index] , 2)
```

```python
print(f' The similarity score is {score} and the best similar document is having index {index} and the sentence similar to the query is {document[index]}')
```
