```python
import zipfile
zip_ref = zipfile.ZipFile('/content/energy-efficiency-data-set.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/energy_efficiency_data.csv')
```

```python
df.head()
```

```python
x = df.drop(['Heating_Load' , 'Cooling_Load'], axis = 1)
y1 = df['Heating_Load']
y2 = df['Cooling_Load']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
import numpy as np
```

```python
y1 = np.array(y1)
y2 = np.array(y2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y1_train , y1_test , y2_train , y2_test = train_test_split(x_scaled , y1 , y2, test_size = 0.2 , random_state = 42)
```

```python
from keras.models import *
from keras.layers import *
```

```python
x_train.shape[1]
```

```python
input = Input(shape = (8,))

d1 = Dense(64 , activation = 'relu')(input)

d2 = Dense(32, activation = 'relu')(d1)

o1 = Dense(1 ,activation = 'linear')(d2)
o2 = Dense(1 ,activation = 'linear')(d2)

model = Model(inputs = input , outputs = [o1 , o2])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=[['mae'] , ['mae']])
```

```python
model.fit(x_train, [y1_train , y2_train] , validation_data = (x_test ,[ y1_test , y2_test]) , epochs = 151 , batch_size = 4 , verbose = 1)
```

```python
model.evaluate(x_test , [y1_test , y2_test])
```
