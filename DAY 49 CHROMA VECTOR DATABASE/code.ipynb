```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
loader = PyPDFLoader(
    'thebook.pdf'
)
```

```python
doc = loader.load()
```

```python
len(doc)
```

```python
doc = doc[0:48]
```

```python
len(doc)
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1300 ,
    chunk_overlap = 0
)
```

```python
doc_split = splitter.split_documents(doc)
```

```python
len(doc_split)
```

```python
chroma_store = Chroma(
    embedding_function=GoogleGenerativeAIEmbeddings(model = 'gemini-embedding-2-preview') , 
    persist_directory='chroma_database' , 
    collection_name='sample'
)
```

```python
doc_split[41]
```

```python
chroma_store.add_documents(doc_split)
```

```python
len(chroma_store.get(include=['embeddings'])['embeddings'][0])
```

```python
ml_similarity = chroma_store.similarity_search(
    query='what is machine learning' , 
    k = 3
)
```

```python
ml_similarity[1].page_content
```

```python
print(ml_similarity[1].page_content)
```

```python
nb_simi_score = chroma_store.similarity_search_with_score(
    query = 'what is naive bayes' , 
    k = 2
)
```

```python
print(nb_simi_score[0][0].page_content)
```
