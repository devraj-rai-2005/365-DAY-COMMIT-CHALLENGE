```python
ct1 = ColumnTransformer(transformers =[
    ('ct1_Exter_Qual', OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , [4] ) ,
    ('ct1_Bsmt_Qual', OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , [5] ) ,
    ('ct1_Heating_QC', OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , [8] ) ,
    ('ct1_Kitchen_Qual', OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , [12] ) ,
    ('ct1_Fireplace_Qu', OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , [15] ) ,
    ('ct1_Garage_Finish', OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , [17] ) ,
] ,remainder = 'passthrough')
```

```python
from sklearn.preprocessing import *
```

```python
ct2 = RandomForestRegressor()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('p1' , ct1) ,
    ('p2' , ct2) ,
])
```
