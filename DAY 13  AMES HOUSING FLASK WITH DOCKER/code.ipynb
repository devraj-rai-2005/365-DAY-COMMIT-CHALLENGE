# Making a Columns Transformer for the Catrgorical Column

```python
from sklearn.compose import ColumnTransformer 
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ct1_encoding = ColumnTransformer(
    transformers = [
        ('ct1_Exter_Qual' ,OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1) , ['Exter_Qual'] ) , 
        ('ct1_Bsmt_Qual' ,OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1) , ['Bsmt_Qual'] ) , 
        ('ct1_Heating_QC' ,OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1) , ['Heating_QC'] ) , 
        ('ct1_Kitchen_Qual' ,OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1) , ['Kitchen_Qual'] ) , 
        ('ct1_Fireplace_Qu' ,OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1) , ['Fireplace_Qu'] ) , 
        ('ct1_Garage_Finish' ,OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1) , ['Garage_Finish'] ) , 
        ], 
    remainder  = 'passthrough'
)
```

```python
from catboost import CatBoostRegressor
```

```python
ct2 = CatBoostRegressor()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('p1' , ct1_encoding ) , 
    ('p2' , ct2 ) , 
])
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x = new_df.drop('SalePrice' , axis = 1)
y = new_df['SalePrice']
```

```python
x_train , x_test  , y_train , y_test = train_test_split(x , y, train_size = 0.2  , random_state = 42)
```

```python
pipe.fit(x_train , y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
pipe.score(x_train , y_train)
```

```python
pipe.score(x_test , y_test)
```

```python
import joblib

joblib.dump(pipe , 'ames_housing_model.pkl')
```
