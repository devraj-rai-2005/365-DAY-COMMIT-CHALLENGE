```python
new_df.head()
```

```python
new_df['Contract'].value_counts()
new_df['Contract'] = new_df['Contract'].str.replace('-' , '_').str.replace(' ' , '_')
new_df['Contract'].value_counts()
```

```python
new_df['City'].value_counts()
new_df['City'] = new_df['City'].str.replace(' ' , '_')
new_df['City'].value_counts()
```
