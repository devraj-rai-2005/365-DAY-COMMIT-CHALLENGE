```python
import zipfile
zip_ref = zipfile.ZipFile('/content/240000-household-electricity-consumption-records.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/household_power_consumption.csv')
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df.drop('index', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['day'] = df['Date'].dt.day
df['month'] = df['Date'].dt.month
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['hour'] = df['Time'].dt.hour
df['minute'] = df['Time'].dt.minute
```

```python
df.drop(['Date', 'Time'] , axis = 1 , inplace = True)
```

```python
df.head(

)
```

```python
df.columns
```

```python
df.columns =['Global_active_power', 'Global_reactive_power', 'Voltage',
       'Global_intensity', 'meter_1', 'meter_2',
       'meter_3', 'day', 'month', 'hour', 'minute']
```

```python
df['meter_1'] = df['meter_1'].astype(float).astype(int)
df['meter_2'] = df['meter_2'].astype(float).astype(int)
df['meter_3'] = df['meter_3'].astype(float).astype(int)
```

```python
df.tail()
```

```python
new_df = df.copy()
```

```python
x = new_df.drop(['meter_1' , 'meter_2' , 'meter_3'],axis = 1)
y1 = new_df['meter_1']
y2 = new_df['meter_2']
y3 = new_df['meter_3']
```

```python
x.shape[1]
```

```python
from sklearn.preprocessing import *
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
import joblib

joblib.dump(scaler , 'ee_24_scaler.pkl')
```

```python
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
```

```python
y1 = np.array(y1)

y2 = np.array(y2)

y3 = np.array(y3)
```

```python
y3
```

```python
x_train , x_test , y1_train , y1_test, y2_train , y2_test , y3_train , y3_test = train_test_split(x_scaled, y1 , y2 , y3 , test_size = 0.2 , random_state = 42)
```

```python
print(x_train.dtype)
print(x_test.dtype)
print(y1_train.dtype, y2_train.dtype, y3_train.dtype)
print(y1_test.dtype, y2_test.dtype, y3_test.dtype)
```

```python
from keras.models import *
from keras.layers import *
```

```python
input = Input(shape = (x_train.shape[1],))

d1 = Dense(1024 , activation = 'relu')(input)

d2 =Dense(1024 , activation = 'relu')(d1)

o1 = Dense(1 , activation = 'linear')(d2)

o2 = Dense(1 , activation = 'linear')(d2)

o3 = Dense(1 , activation = 'linear')(d2)


model = Model(inputs = input , outputs = [o1 , o2 , o3])
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam' , loss = 'mse' , metrics =[ ['mae'] , ['mae'] , ['mae']] )
```

```python
model.fit(x_train , [y1_train , y2_train , y3_train] ,validation_data = ( x_test , [y1_test , y2_test , y3_test]) , epochs = 50 , batch_size = 32 , verbose = 1 )
```

```python
model.save('24_ee_model.keras')
model.save('24_ee_model.h5')
```
