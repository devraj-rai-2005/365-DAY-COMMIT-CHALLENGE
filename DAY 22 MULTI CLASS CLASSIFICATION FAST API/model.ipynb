```python
import pandas as pd
```

```python
df = pd.read_csv('model.csv')
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
ct1 = ColumnTransformer(
    transformers= [
        ('ct1_Workclass' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Workclass']) ,

        ('ct1_Education' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Education']) ,

        ('ct1_Martial_Status' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Martial_Status']) ,

        ('ct1_Occupation' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Occupation']) ,

        ('ct1_Relationship' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Relationship']) ,

        ('ct1_Race' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Race']) ,

        ('ct1_Sex' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Sex']) ,

        ('ct1_Country' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1) , ['Country'])
    ] , 
    remainder = 'passthrough'
)
```

```python
from catboost import CatBoostClassifier
```

```python
ct2 = CatBoostClassifier()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('p1' , ct1 ) , 
    ('p2' , ct2 ) , 
])
```

```python
from sklearn.model_selection import  train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Target'].value_counts()
```

```python
df['Target'] = le.fit_transform(df['Target'])
```

```python
x = df.drop('Target' , axis =1 )
y = df['Target']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 ,random_state = 42)
```

```python
pipe.fit(x_train , y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
pipe.score(x_train , y_train)
```

```python
pipe.score(x_test , y_test )
```
