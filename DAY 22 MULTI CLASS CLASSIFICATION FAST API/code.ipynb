```python
import pandas as pd
```

```python
df = pd.read_csv('model.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
new_df = df.copy()
```

```python
new_df.columns
```

```python
from sklearn.preprocessing import *
```

```python
ol = OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1)
```

```python
new_df['Workclass'] = ol.fit_transform(new_df[['Workclass']]).astype(int)
new_df['Education'] = ol.fit_transform(new_df[['Education']]).astype(int)
new_df['Martial_Status'] = ol.fit_transform(new_df[['Martial_Status']]).astype(int)
new_df['Occupation'] = ol.fit_transform(new_df[['Occupation']]).astype(int)
new_df['Relationship'] = ol.fit_transform(new_df[['Relationship']]).astype(int)
new_df['Race'] = ol.fit_transform(new_df[['Race']]).astype(int)
new_df['Sex'] = ol.fit_transform(new_df[['Sex']]).astype(int)
new_df['Country'] = ol.fit_transform(new_df[['Country']]).astype(int)
```

```python
le = LabelEncoder()
```

```python
new_df['Target'] = le.fit_transform(new_df['Target'])
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = new_df.drop('Target' , axis = 1)
y = new_df['Target']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 ,random_state = 42)
```

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier()
```

```python
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_train , y_train)
```

```python
model.score(x_test , y_test)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
confusion_matrix(y_test , y_pred)
```
