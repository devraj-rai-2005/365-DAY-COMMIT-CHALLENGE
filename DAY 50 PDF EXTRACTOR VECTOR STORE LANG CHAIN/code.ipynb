```python
from langchain_community.document_loaders import PyPDFLoader
from dotenv import load_dotenv
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
```

```python
load_dotenv()

model = GoogleGenerativeAIEmbeddings(model = 'gemini-embedding-2-preview')
```

```python
loader = PyPDFLoader('The_Verdict.pdf')
```

```python
doc = loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500 , 
    chunk_overlap = 0
)
```

```python
doc_split = splitter.split_documents(doc)
```

```python
len(doc_split)
```

```python
len(doc)
```

```python
chroma_store = Chroma(
    embedding_function=model , 
    persist_directory='chroma_database' , 
    collection_name='sample'
)
```

```python
chroma_store.add_documents(doc_split)
```

```python
len(chroma_store.get(include = ['embeddings'])['embeddings'][0])
```
