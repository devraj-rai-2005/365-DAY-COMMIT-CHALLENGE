```python
from sklearn.preprocessing import LabelEncoder
```

```python
x = new_df.drop('default', axis =1 )
y = new_df['default']
```

```python
x.head()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import *
```

```python
ct1 = ColumnTransformer(
    transformers = [
        ('ct1_SEX' , OneHotEncoder(sparse_output = False , drop= 'first' , dtype = int) , ['SEX']) ,
        ('ct1_EDUCATION' , OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1 , dtype = int) , ['EDUCATION']) , 
        ('ct1_MARRIAGE' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1 , dtype = int) , ['MARRIAGE'])
        ] ,
    remainder = 'passthrough'

)
```

```python
from imblearn.combine import SMOTETomek

sampling = SMOTETomek(random_state = 42)

ct2 = sampling
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier 
```

```python
model = CatBoostClassifier()
```

```python
ct3 = model
```

```python
from imblearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('p1' , ct1) , 
    ('p3' , ct3),
])
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size = 0.2 , random_state = 42)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test ,y_pred))
```

```python
confusion_matrix(y_test , y_pred)
```
