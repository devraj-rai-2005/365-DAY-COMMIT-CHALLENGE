```python
from langchain_community.retrievers import WikipediaRetriever
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
```

```python
load_dotenv()

embedding_model = GoogleGenerativeAIEmbeddings(model = 'gemini-embedding-2-preview')
```

```python
retriver = WikipediaRetriever(
    top_k_results=4 , 
    lang = 'en'
)
```

```python
query = 'Anal Sex'
```

```python
doc = retriver.invoke(query)
```

```python
for i in range(len(doc)):
    print('length of Document {i} : ' , len(doc[i].page_content))
```

```python
split = RecursiveCharacterTextSplitter(
    chunk_size = 1500 , 
    chunk_overlap = 0 
)
```

```python
doc_split = split.split_documents(doc)
```

```python
len(doc_split)
```

```python
for i in range(len(doc_split)):
    print('length of Document {i} : ' , len(doc_split[i].page_content))
```

```python
chroma = Chroma(
    embedding_function= embedding_model , 
    collection_name = 'sample' , 
    persist_directory='chroma_directory'
)
```

```python
chroma.add_documents(doc_split)
```

```python
print(chroma.get()['documents'][0])
```
